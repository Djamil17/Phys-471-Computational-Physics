```python
batch=pd.read_parquet('/Users/dlakhdar/Downloads/batch_1.parquet')
sensors=pd.read_csv('/Users/dlakhdar/Downloads/sensor_geometry.csv')
```

```python
norm= np.sqrt((sensors.x)**2+(sensors.y)**2+(sensors.z)**2)

for i in ['x_','y_','z_']:
    sensors[i]=sensors.x/norm

def convert_cartesian_to_spherical(x,y,z):
    """
    Takes three pandas columns x,y,z each a cartesian component and spits out two columns θ,ϕ 
    """
    
    θ=np.arccos(z)
    ϕ=np.arcsin(x)/np.cos(θ)
    
    return (θ,ϕ)

zenith, azimuth=convert_cartesian_to_spherical(sensors.x_,sensors.y_,sensors.z_)
sensors['zenith']=zenith
sensors['azimuth']=azimuth
```
